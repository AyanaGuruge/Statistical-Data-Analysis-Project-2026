### Step 2: Identify and Handle Missing Values

```python
# Check for missing values represented as '?'
print("Missing values (represented as '?'):")
for col in df.columns:
    missing_count = (df[col] == '?').sum()
    if missing_count > 0:
        print(f"  {col}: {missing_count} missing values")

# Display summary of missing data
print("\nTotal missing values by column:")
print(df.isnull().sum())

# Check data quality
print("\nData Quality Report:")
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"Duplicate rows: {df.duplicated().sum()}")
```

### Step 3: Data Type Conversion and Cleaning

```python
# Replace '?' with NaN for proper handling
df_clean = df.replace('?', np.nan)

# Convert demographic variables to appropriate types
df_clean['kohde'] = pd.to_numeric(df_clean['kohde'], errors='coerce')
df_clean['jasen'] = pd.to_numeric(df_clean['jasen'], errors='coerce')
df_clean['pvknro'] = pd.to_numeric(df_clean['pvknro'], errors='coerce')
df_clean['sp'] = pd.to_numeric(df_clean['sp'], errors='coerce')
df_clean['IKAL1'] = pd.to_numeric(df_clean['IKAL1'], errors='coerce')
df_clean['ASALUE'] = pd.to_numeric(df_clean['ASALUE'], errors='coerce')

# Activity variables - convert to numeric, handling time format data
for col in ['A1', 'A2', 'A3', 'A4', 'A5']:
    # Try to convert directly to numeric first
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

print("Data types after conversion:")
print(df_clean.dtypes)
print("\nMissing values per column:")
print(df_clean.isnull().sum())
print("\nData shape:", df_clean.shape)
```

### Step 4: Add Variable Labels and Summary Statistics

```python
# Create variable labels dictionary
variable_labels = {
    'kohde': 'Household ID',
    'jasen': 'Member ID',
    'pvknro': 'Day of Week (1=Weekday, 2=Weekend)',
    'sp': 'Sex (1=Male, 2=Female)',
    'IKAL1': 'Age Group',
    'ASALUE': 'Living Environment (1=City, 2=Municipality, 3=Rural)',
    'A1': 'Working (minutes)',
    'A2': 'Sleeping (minutes)',
    'A3': 'Reading (minutes)',
    'A4': 'Dining at Restaurant (minutes)',
    'A5': 'Visiting Library (yes/no or minutes)'
}

# Display summary statistics
print("=" * 80)
print("DESCRIPTIVE STATISTICS")
print("=" * 80)
print(df_clean.describe())

print("\n" + "=" * 80)
print("VARIABLE INFORMATION")
print("=" * 80)
for col, label in variable_labels.items():
    print(f"{col:10} | {label}")

print("\n" + "=" * 80)
print("MISSING DATA SUMMARY")
print("=" * 80)
missing_summary = pd.DataFrame({
    'Column': df_clean.columns,
    'Missing Count': df_clean.isnull().sum(),
    'Missing %': (df_clean.isnull().sum() / len(df_clean) * 100).round(2)
})
print(missing_summary)
```
